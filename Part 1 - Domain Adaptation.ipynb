```python
Fast_Off="/Users/student/Fast_Offensive_Training.csv"
```

```python
Fast_Off_pd = pd.read_csv(Fast_Off, encoding ='ISO-8859-1')
XFast=pd.DataFrame(data=Fast_Off_pd, columns=['Tweet'])
```

```python
XFast['Tweet'] = XFast['Tweet'].replace('[^a-zA-Z ]', '', regex=True)
```

```python
XF = XFast.apply(lambda x: x.astype(str).str.lower())
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
Lemmatizer = WordNetLemmatizer()
tokenizer= TweetTokenizer()
```

```python
def lemmatize_text(Tweet):
    return [Lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(Tweet) if not w in set(stopwords.words('english'))]
```

```python
XF['Tweet']=XF.Tweet.apply(lemmatize_text)
```

```python
fast = XF['Tweet']
```

```python
fast
```

```python
frequencies = Counter (word for sentence in fast for word in sentence)
```

```python
n=5000
for word, frequency in frequencies.most_common()[:-n-1:-1]:
    print (word, frequency)
```
